```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Cargar el archivo CSV
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Extraer y filtrar los saldos de cuentas de ahorro para mujeres
# Se busca identificar patrones, rangos frecuentes y posibles concentraciones de valor. 
depositos_mujeres = df['SALDO_CTA_AHORRO_MUJERES'].dropna()
depositos_mujeres = depositos_mujeres[depositos_mujeres > 0]
```

```python
# Eliminar outliers (percentil 99)
# Eliminamos outliers para excluir valores extremos que podrian distorsionar el analisis de la distribucion, mejorando la claridad del histograma.
p99_mujeres = depositos_mujeres.quantile(0.99)
depositos_mujeres_filtrados = depositos_mujeres[depositos_mujeres <= p99_mujeres]
```

```python
# Convertir a millones para facilitar la lectura
depositos_mujeres_filtrados = depositos_mujeres_filtrados / 1_000_000
```

```python
# Crear histograma solo para mujeres
plt.figure(figsize=(8,6))
plt.hist(depositos_mujeres_filtrados, bins=30, color='#FF69B4', edgecolor='black', alpha=0.8)
plt.title('Histograma de Saldos de Cuentas de Ahorro - Mujeres (sin outliers)')
plt.xlabel('Saldo de Cuenta de Ahorro (Millones)')
plt.ylabel('Frecuencia')
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.2f}'))
plt.tight_layout()
plt.show()
```

```python
# La gran mayoria de los saldos se concentran en el rango mas bajo (cercano a 0).
# Hay una fuerte caida en la frecuencia a medida que aumentan los saldos.
# La distribucion es altamente sesgada a la derecha, es decir, hay pocas mujeres con saldos altos.
# Se observa que mas de 30,000 mujeres tienen saldos muy bajos.
```

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Cargar datos
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Extraer y filtrar los saldos de cuentas de ahorro para hombres
# Se busca identificar patrones, rangos frecuentes y posibles concentraciones de valor. 
depositos_hombres = df['SALDO_CTA_AHORRO_HOMBRES'].dropna()
depositos_hombres = depositos_hombres[depositos_hombres > 0]
```

```python
# Eliminar outliers (percentil 99)
# Eliminamos outliers para excluir valores extremos que podrian distorsionar el analisis de la distribucion, mejorando la claridad del histograma.
p99_hombres = depositos_hombres.quantile(0.99)
depositos_hombres_filtrados = depositos_hombres[depositos_hombres <= p99_hombres]
```

```python
# Convertir a millones para facilitar la lectura
depositos_hombres_filtrados = depositos_hombres_filtrados / 1_000_000
```

```python
# Crear histograma solo para hombres
plt.figure(figsize=(8,6))
plt.hist(depositos_hombres_filtrados, bins=30, color='#87CEFA', edgecolor='black', alpha=0.8)
plt.title('Histograma de Saldos de Cuentas de Ahorro - Hombres (sin outliers)')
plt.xlabel('Saldo de Cuenta de Ahorro (Millones)')
plt.ylabel('Frecuencia')
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.2f}'))
plt.tight_layout()
plt.show()
```

```python
# La tendencia es muy similar a la de las mujeres: alta concentracion en saldos bajos.
# Tambien hay una fuerte caida en la frecuencia a medida que se incrementan los saldos.
# Sin embargo, se nota una mayor dispersion en los saldos altos en comparacion con las mujeres, con algunos hombres alcanzando valores mas elevados.
# Tambien mas de 30,000 hombres tienen saldos bajos, lo que indica un comportamiento masivo similar en la base.
# los hombres presentan una mayor presencia en los niveles de saldo alto, lo que sugiere mayor capacidad de acumulacion financiera, posiblemente asociada a diferencias en ingreso, acceso a oportunidades economicas o uso de instrumentos financieros.
```
