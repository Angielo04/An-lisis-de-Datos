```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar datos
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Calcular totales por género
cuentas_mujeres = df['NRO_CTA_AHORRO_MUJERES'].sum()
cuentas_hombres = df['NRO_CTA_AHORRO_HOMBRES'].sum()
```

```python
# Crear gráfico de pastel
# Se elige graficos de tipo pastel para representar la proporcion de participacion de cuentas de ahorro por genero.
labels = ['Mujeres', 'Hombres']
sizes = [cuentas_mujeres, cuentas_hombres]
```

```python
# asigna etiquetas a cada segmento con 'labels',
# muestra el porcentaje de cada segmento con un decimal ('%1.1f%%')
# rotamos el grafico 90 grados para que el primer segmento inicie desde la parte superior.
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.set_title('Cantidad de Cuentas de Ahorro por Género')
plt.show()
```

```python
# los hombres poseen el 52.8% de las cuentas de ahorro, mientras que las mujeres tienen el 47.2%.
# Esto revela una brecha de 5.6 puntos porcentuales a favor de los hombres.
# Esto indica una ligera predominancia masculina en la apertura de cuentas de ahorro. 
# Para mejorar la inclusion financiera, es crucial fomentar la participacion de las mujeres en el sistema de ahorro.
```

```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Caragar datos
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Calcular total por tipo de crédito (sumando mujeres y hombres)
# Se calcula los totales por tipo de credito para identificar los productos crediticios mas utilizados.
total_consumo = df['MONTO_CREDITO_CONSUMO_MUJERES'].sum() + df['MONTO_CREDITO_CONSUMO_HOMBRES'].sum()
total_vivienda = df['MONTO_CREDITO_VIVIENDA_MUJERES'].sum() + df['MONTO_CREDITO_VIVIENDA_HOMBRES'].sum()
total_microcredito = df['MONTO_MICROCREDITO_MUJERES'].sum() + df['MONTO_MICROCREDITO_HOMBRES'].sum()

# Mostrar resultados
total_consumo, total_vivienda, total_microcredito
```

```python
# Preparar datos para grafico de pastel
# Se facilita la representacion visual de la proporcion de cada tipo de credito.
labels = ['Crédito de Consumo', 'Crédito de Vivienda', 'Microcrédito']
sizes = [total_consumo, total_vivienda, total_microcredito]
colors = ['#ff9999','#66b3ff','#99ff99']

# Crear gráfico de pastel
plt.figure(figsize=(8,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Total de Saldos por Tipo de Crédito')
plt.axis('equal')  # Para que sea un círculo perfecto
plt.show()
```

```python
# Credito de Consumo: Representa el 78.9% del total, indicando que la mayoia de los saldos de credito se destinan al consumo personal.
# Credito de Vivienda: Constituye el 14.5% del total, reflejando una parte significativa pero menor participacion, posiblemente debido a requisitos mas estrictos.
# Microcredito: Comprende el 6.5% del total, sugiriendo un uso limitado de este tipo de credito, a pesar de su importancia para pequeñas empresas y emprendedores.
# Este panorama sugiere que aun hay desigualdad en el acceso a creditos productivos, lo cual limita la movilidad economica y el desarrollo a largo plazo, especialmente entre los sectores tradicionalmente excluidos. 
```
