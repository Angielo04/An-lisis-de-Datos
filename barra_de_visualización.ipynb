```python
# Importar las librerias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Agrupar los datos por DEPARTAMENTO y sumar el numero de cuentas de ahorro activas
# Nos permite identificar que departamentos tienen mayor actividad en cuentas de ahorro y facilita el analisis por region.
cuentas_activas = df.groupby('DEPARTAMENTO')['NRO_CTA_AHORRO_ACTIVAS'].sum().reset_index()
```

```python
# Ordenar los datos de mayor a menor para una mejor visualizacion
# Realizamos la organizacion de las cifras para identificar rapidamente los valores mas altos y facilitar la comparacion de los elementos
cuentas_activas = cuentas_activas.sort_values(by='NRO_CTA_AHORRO_ACTIVAS', ascending=False)
```

```python
# Crear el grafico de barras con Seaborn
# Se elige iniciar con grafico de barras, este nos permite realizar una comparacion general con esta base de datos, partiendo desde un punto de vista general.
plt.figure(figsize=(12, 6))
sns.barplot(data=cuentas_activas, x='DEPARTAMENTO', y='NRO_CTA_AHORRO_ACTIVAS', hue='DEPARTAMENTO', palette='viridis', legend=False)


# Mejorar el diseño del grafico
# Se organiza la informacion en el grafico para visualizar de manera mas sencilla y concisa.
plt.xticks(rotation=45, ha='right')
plt.title('Número total de cuentas de ahorro activas por departamento')
plt.xlabel('Departamento')
plt.ylabel('Número de cuentas activas')
plt.tight_layout()

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Mostrar el grafico
plt.show()
```

```python
# La grafica de barras muestra el numero total de cuentas de ahorro activas agrupadas por Departamento.
# Los datos ordenados de mayor a menor permiten identificar rapidamente las regiones con mayor y menor actividad financiera.
# Antioquia, Bogota y Valle del Cauca lideran con los valores mas altos, mientras que departamentos como Vaupes, Guainia y Amazonas presentan los valores mas bajos.
# Esta comparacion resalta la concentracion del uso de productos financieros basicos como las cuentas de ahorro en regiones densamente pobladas.
```

```python
# Importar las librerias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Crear un nuevo DataFrame con la suma total de cada tipo de producto por genero
# Se realiza para analizar el comportamiento y la distribucion del uso de productos financieros por genero
datos_genero = pd.DataFrame({
    'Producto': ['Cuentas de Ahorro', 'Cuentas de Ahorro', 'Crédito Consumo', 'Crédito Consumo'],
    'Género': ['Mujeres', 'Hombres'] * 2,
    'Cantidad': [
        df['NRO_CTA_AHORRO_MUJERES'].sum(), df['NRO_CTA_AHORRO_HOMBRES'].sum(),
        df['NRO_CREDITO_CONSUMO_MUJERES'].sum(), df['NRO_CREDITO_CONSUMO_HOMBRES'].sum()
        ]
})
```

```python
# Crear el grafico de barras comparativo
# Se elige grafico de barras ya que facilita la comparacion directa entre categorias
plt.figure(figsize=(10, 6))
sns.barplot(data=datos_genero, x='Producto', y='Cantidad', hue='Género', palette='pastel')

# Mejorar el diseño del grafico
# Se organiza para mejorar su visualizacion e interpretacion
plt.title('Comparación entre hombres y mujeres en productos financieros')
plt.xlabel('Tipo de producto financiero')
plt.ylabel('Cantidad de productos')
plt.xticks(rotation=45)

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# En ambos tipos de productos financieros, los hombres tienen mayor cantidad de cuentas y creditos en comparacion con las mujeres.
# Esto podria indicar una mayor inclusion financiera de los hombres en estos productos especificos.
# Existe una brecha de genero en cuentas de ahorro a favor de los hombres, esto sugiere una menor inclusion financiera de las mujeres en productos de ahorro.
# En creditos de consumo, la equidad es mucho mayor, lo que podria indicar un acceso mas igualitario a este tipo de producto.
```

```python
# Importar las librerias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
df = pd.read_csv("C:\\Users\\anggl\\OneDrive\\Escritorio\\trabajo python\\Financiera_clean.csv")
```

```python
# Crear un nuevo DataFrame con la suma total de cada tipo de producto por genero.
# Se realiza para analizar el comportamiento y la distribucion del uso de productos financieros por genero.
datos_genero = pd.DataFrame({
    'Producto': ['Crédito Vivienda', 'Crédito Vivienda', 'Microcrédito', 'Microcrédito'],
    'Género': ['Mujeres', 'Hombres'] * 2,
    'Cantidad': [
        df['NRO_CREDITO_VIVIENDA_MUJERES'].sum(), df['NRO_CREDITO_VIVIENDA_HOMBRES'].sum(),
        df['NRO_MICROCREDITO_MUJERES'].sum(), df['NRO_MICROCREDITO_HOMBRES'].sum()
    ]
})
```

```python
# Crear el grafico de barras comparativo
# Se elige grafico de barras ya que facilita la comparacion directa entre categorias
plt.figure(figsize=(10, 6))
sns.barplot(data=datos_genero, x='Producto', y='Cantidad', hue='Género', palette='pastel')

# Mejorar el diseño del grafico
# Se organiza para mejorar su visualizacion e interpretacion
plt.title('Comparación entre hombres y mujeres en productos financieros')
plt.xlabel('Tipo de producto financiero')
plt.ylabel('Cantidad de productos')
plt.xticks(rotation=45)

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# En el caso del credito de vivienda, los hombres tienen una mayor cantidad de creditos en comparacion con las mujeres. Esto podria indicar una mayor participacion de los hombres en la adquisicion de viviendas.
# En el caso del microcredito, las mujeres tienen una mayor cantidad de creditos en comparacion con los hombres. Esto sugiere que las mujeres estan mas involucradas en el acceso a microcreditos, posiblemente para emprendimientos, pequeñas empresas o en sectores vulnerables.
# La inclusion financiera sigue siendo desigual en productos tradicionales como ahorro y vivienda.
# Las mujeres han ganado terreno en productos orientados a la base de la piramide, como el microcredito.
# La equidad en creditos de consumo es un indicador positivo, pero persiste la necesidad de politicas que fomenten el acceso femenino a productos financieros de largo plazo y ahorro.
```
